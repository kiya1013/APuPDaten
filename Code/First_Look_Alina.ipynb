```python
df["timestamp"] = pd.to_datetime(df["timestamp"])
```

```python
df = df[df["DeviationID ValueY"] != 0]
```

```python
df.groupby("DeviationID ValueY").max().drop(columns=["timestamp", "CuStepNo ValueY"])
```

```python
df.groupby("DeviationID ValueY").min().drop(columns=["timestamp", "CuStepNo ValueY"])
```

Idee: Schauen, was sind die Min- und Max-Werte bei `DeviationID ValueY`. Alles was drüber / darunter ist, ist Anomalie.

```python
df["Anomalie"] = (df["DeviationID ValueY"] != 1)
```

```python
df["Anomalie"] = df["Anomalie"].astype(int)
```
